```python
from autogluon.tabular import TabularPredictor
```

```python
import pandas as pd
```

```python
opt_train = pd.read_csv("optimized_train.csv")
```

```python
predictor_new_hpo = TabularPredictor(label = "Machine failure").fit(train_data = opt_train, time_limit = 1000, presets = "best_quality",hyperparameters = "default", hyperparameter_tune_kwargs = "random" )
```

```python
model = TabularPredictor.load("AutogluonModels/ag-20230617_063449/")
```

```python
model.fit_summary()
```

```python
test = pd.read_csv("optimized_test.csv")
probabilities = model.predict_proba(test)
```

```python
submission = pd.read_csv("sample_submission.csv")
submission.head()
```

```python
probs = probabilities[1].values
```

```python
submission["Machine failure"] = probs
```
